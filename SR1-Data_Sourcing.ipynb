# Informal Settlements Dataset Analysis

## Dataset Overview
The dataset selected was sourced from the **Humanitarian Data Exchange (HDX)**, a trusted platform managed by the United Nations. The dataset focuses on informal settlements and provides key information on settlement names, population sizes, geographic coordinates, and shelter types, among other variables.

## Justification for Dataset Selection
This dataset is suitable for the analysis of informal settlements because it contains several relevant columns:

- **Settlement_Name**: This column provides a unique identifier for each informal settlement, allowing us to study patterns and differentiate between various locations.
  
- **Population**: This column offers insight into the size of each settlement, which is essential for analyzing density, overcrowding, and resource needs.

Additionally, the data's source ensures its credibility, as it is published and regularly updated by the United Nations Office for the Coordination of Humanitarian Affairs (OCHA), a well-recognized organization in the field of humanitarian data.

## Data Completeness
The dataset is comprehensive, covering a wide range of settlements. By inspecting missing values, we can ensure the completeness and reliability of the dataset for our analysis.
